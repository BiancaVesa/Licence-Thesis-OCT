# Visualize OCTs and encodings

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from ipywidgets import IntSlider
import glob
import pickle
import tensorflow
import numpy as np
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
def show_octs_and_encodings(patient_nr, visit_nr, eye):
    images = glob.glob("D:/AN4/Licenta/Datasets/AMD/UDF Volume/Pacient "+ str(patient_nr) + "/Vizita " + str(visit_nr) + "*/" + eye + "/*.jpg")
    path = glob.glob("D:/AN4/Licenta/Datasets/AMD_encoded/Patient_" + str(patient_nr) + "_visit_" + str(visit_nr) + "_*_eye_" + eye + "_*.txt")
    
    with open(path[0], 'rb') as tensor_file:
        tensor = pickle.load(tensor_file)
    
    x_widget = IntSlider(min=1, max=25, step=1, value=1)
    
    index_image_name = path[0].index("Patient")
    def show(x):
        image = plt.imread(images[x])
        plt.title(path[0][index_image_name:])
        plt.imshow(image, cmap='gray')
    interact(show, x=x_widget)
    plt.imshow(tensor)
```

```python
show_octs_and_encodings(35, 2, "OD")
```

```python
show_octs_and_encodings(74, 5, "OS")
```
